```python
import pandas as pd
```

```python
df = pd.read_excel('number-of-sheeps-by-emirate-from-2005-2018-ds-.xlsx')
df
```

```python
df.shape
```

```python
df.columns
```

```python
df.dropna()
df
```

```python
Q1 = df[['Year','Count']]
Q1 = Q1.groupby("Year").sum()
Q1 = Q1.sort_values("Count")
Q1
```

```python
Q1["Count"].plot.pie(title="Sheeps present per Year",autopct="%.2f%%")
```

```python
Q2 = df[(df["Emirates_en"]=='Abu Dhabi')|(df["Emirates_en"]=='Dubai')|(df["Emirates_en"]=='Ajman')]
Q2 = Q2.groupby(['Emirates_en','Year','Count']).sum()
Q2
```

```python
Q2.plot()
```

```python
Q3 = df[df['Year']==2006].groupby(['Year','Emirates_en']).sum()
Q3
```

```python
Q3["Count"].plot()
```

```python
Q4 = df["Count"].mean()
Q4 = df.groupby(['Emirates_en','Count']).mean()
Q4
```

```python
Q4.plot()
```

```python
Q7 = df[df['Year']==2016].groupby(['Year','Emirates_en']).sum()
Q7
```

```python
Q7["Count"].plot()
```
